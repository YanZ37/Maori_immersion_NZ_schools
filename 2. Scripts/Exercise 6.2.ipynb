#### Create a correlation heatmap using matplotlib

```python
selected_vars = ['School Decile', 'Total Roll', 'Māori', 'European/ Pākehā','Immersion Level 1','Immersion Level 2',
                 'Immersion Level 3','Immersion Level 4A','Immersion Level 4B', 'Immersion Level 5','Immersion Taha Māori','Immersion None']
```

```python
# Make sure everything is numeric
corr_matrix = df[selected_vars].apply(pd.to_numeric, errors='coerce').corr()

fig, ax = plt.subplots(figsize=(10, 10))
im = ax.matshow(corr_matrix.to_numpy(), cmap='viridis')

ax.set_xticks(np.arange(corr_matrix.shape[1]))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='left', fontsize=12)

ax.set_yticks(np.arange(corr_matrix.shape[0]))
ax.set_yticklabels(corr_matrix.index, fontsize=12)

cbar = fig.colorbar(im, ax=ax)
cbar.ax.tick_params(labelsize=12)

ax.set_title('Correlation Matrix', fontsize=16, pad=12)
fig.tight_layout()
plt.savefig(r"C:\Users\sanch\Documents\CareerFoundry\Achievement 6\3. Visualisations\correlation_matrix.png")
plt.show()
```

#### Categorical Plots:

Creating a categorical variable that splits the "Decile" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the decile ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['School Decile'], bins = 30, kde = True)
fig.tight_layout()
plt.savefig(r"C:\Users\sanch\Documents\CareerFoundry\Achievement 6\3. Visualisations\histplot.png")
plt.show()
```

**Observations:** 
The distribution of School Decile is even across schools. 

```python
df.loc[df['School Decile'] < 4, 'Decile Band'] = 'Low decile'
df.loc[(df['School Decile'] >= 4) & (df['School Decile'] < 8), 'Decile Band'] = 'Middle decile'
df.loc[df['School Decile'] >= 8, 'Decile Band'] = 'High decile'
```
